Importanto dados

```python
import pandas as pd
```

Coletando os dados

```python
dados_2019 = pd.read_csv('análise de dados/emprestimos-20192.csv')
dados_2020 = pd.read_csv('análise de dados/emprestimos-20201.csv')
```

Concatenando os dataframes

```python
emprestimos_biblioteca = pd.concat([dados_2019,dados_2020],ignore_index=True)
emprestimos_biblioteca
```

Verificando duplicatas

```python
emprestimos_biblioteca.value_counts()
```

Excluindo duplicatas

```python

emprestimos_biblioteca = emprestimos_biblioteca.drop_duplicates()
```

```python
emprestimos_biblioteca.value_counts()
```
